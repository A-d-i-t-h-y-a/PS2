Oil spill incidents occur due to various reasons and contaminate
the water body, forming oil sludge in the sea and along the coast.
The spill aï¬€ects severely the seagrasses, coral reefs, aquaculture,
desalination plants, etc. in the marine environment and distresses
the economy, including tourism activities.

An Oil Spill Index (OSI = (B3+B4)/B2) was developed and applied to
Sentinel-2 optical satellite data of European Space Agency (ESA) to
**map of marine oil spills** using the spectral absorption characters of
spectral bands of the Sentinel-2


This is the analysis based on incident happened at Mauritius, Indian Ocean, on August 06, 2020
for a basic analysis, we need two datasets.... before and after the happening of the incident
I have chosen the satellite image of 2019-03-30(before) and 2020-08-11(after)

```python
%matplotlib inline
# This is a command specific to Jupyter notebooks
# that allows matplotlib plots to be displayed directly within the notebook instead of in a separate window.
import datacube
from deafrica_tools.plotting import rgb, display_map
import odc.algo
import matplotlib.pyplot as plt
import numpy as np

# -------------------------------------------------------------------------------------------------------------
dc = datacube.Datacube(app="oil_spill_index")


# specifying the lattitude longitude range
lat_range = (-20.446074, -20.4246)
lon_range = (57.726288, 57.752895)


# --------------------------------------------------------------------------------------------------------------

# This is a function provided by the datacube package for loading data from a datacube.
# It retrieves satellite data based on the specified parameters.


ds = dc.load(product="s2a_sen2cor_granule",
                 measurements=["red","green","blue", "nir"],
             x=lon_range,
             y=lat_range,
             output_crs='EPSG:6933',
             resolution=(-10, 10))


# EPSG:6933 refers to a specific coordinate reference system (CRS)
print(ds)
```

```python
display_map(x=lon_range, y=lat_range)
```

```python
# Load the data from the Datacube
dataset = ds

rgb(ds, bands=['red','green', 'blue',], index=[0, 1])

# index=[0, 1]: This parameter specifies the time index or indices of the data to be used for creating 
# the RGB composite. The value [0, 1] indicates that the first and second time slices of the data will be used.
```

```python
# convert dataset to float32 datatype so no-data values are set to NaN
dataset =  odc.algo.to_f32(dataset)
# Below are the reasons to convert into float 32.....
# Ensuring consistent data types
# Compatibility with algorithms
# Enhanced precision
```

OSI = (B3 + B4) / B2 => (green+red)/blue
This is the formula , we gonna be using in the to analyse the oil spill index

```python
# Calculate the components that make up the oil spill index (OSI)
osi=(dataset.green + dataset.red) / dataset.blue
# Calculate OSI and store it as a measurement in the original dataset
osi=osi/3;
print(osi)
```

```python
plt.figure(figsize=(8, 8))
osi.plot(col='time', cmap='hot', vmin=0.008, vmax=1, col_wrap=3)

plt.show()
```

```python
# calculating the difference before and after the incident happened so we can to know the oil spills 
res_start = osi.sel(time='2019').mean(dim='time')
res_end = osi.sel(time='2020').mean(dim='time')
res_diff = res_end - res_start
```

```python
# plotting the diference map 
plt.figure(figsize=(8, 8))
res_diff.plot(cmap='hot', vmin=0, vmax=0.8, col_wrap=3)
plt.show()
```

From the above figure, we can know that yellow bright region indicates the oil spill..
therefore, we got to know the area of oil spill....!
