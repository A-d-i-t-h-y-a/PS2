```python

import datacube
from deafrica_tools.plotting import rgb, display_map
import datacube
import odc.algo
import matplotlib.pyplot as plt
from datacube.utils.cog import write_cog

from deafrica_tools.plotting import display_map, rgb
dc = datacube.Datacube(app="04_Plotting")
#15.85828652, 80.78694696
#15.75418332, 81.02203692
lat_range = (15.75418332, 15.85828652)
lon_range = (80.78694696, 81.02203692)
time_range = ('2022-01-15', '2023-02-15')
#display_map(x=lon_range, y=lat_range)
ds = dc.load(product="s2a_sen2cor_granule",
                 measurements=["B04_10m","B03_10m","B02_10m", "B08_10m","B11_20m","B12_20m"],
             x=lon_range,
             y=lat_range,
             time=time_range,
             output_crs='EPSG:6933',
             resolution=(-30, 30))

print(ds)
```

```python
display_map(lon_range, lat_range)
```

```python
rgb(ds,bands=['B04_10m', 'B03_10m', 'B02_10m'],index=[0,1])
```

```python
dataset=ds
import numpy as np
# convert dataset to float32 datatype so no-data values are set to NaN
dataset =  odc.algo.to_f32(dataset)

# Calculate the ENDISI index
mndwi=(dataset.B03_10m-dataset.B11_20m)/(dataset.B03_10m+dataset.B11_20m)
print(mndwi)

print("hhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhh")

swir_diff=(dataset.B11_20m)/(dataset.B12_20m)
print(swir_diff)

print("hhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhhh")

alpha=(2*(np.mean(dataset.B02_10m)))/((np.mean(swir_diff))+(np.mean(mndwi*mndwi)))
print(alpha)
```

```python
ENDISI=(dataset.B02_10m-alpha*(swir_diff+(mndwi*mndwi)))/(dataset.B02_10m+alpha*(swir_diff+(mndwi*mndwi)))
print(ENDISI)
```

```python
"""dataset.ENDISI.plot(
    col='year',
    vmin=-.75, 
    vmax=0.75, 
    cmap='RdBu',
    figsize=(10, 5),
    robust=True
);"""
plt.figure(figsize=(8, 8))
ENDISI.plot(col='time', vmin=0, vmax=1, col_wrap=3)
plt.show()
```

```python
ENDISI.plot.hist(bins=1000, range=(-1,1), facecolor='gray', figsize=(10, 4))
plt.title('ENDISI Histogram');
```

```python
# Create a query
baseline_year=2022
analysis_year=2023
query = {
    'time': (f'{baseline_year}', f'{analysis_year}'),
    'x': lon_range,
    'y': lat_range,
    'resolution': (-20, 20),
    'measurements':['swir_1','swir_2','blue','green','red'],
    'group_by': 'solar_day',
    
}
```

```python
ENDISI = ENDISI.fillna(ENDISI.mean().values)

from skimage.filters import threshold_otsu

threshold = threshold_otsu(ENDISI.values)
print(round(threshold, 2))
```

```python
urban_area = (ENDISI > threshold).astype(int)
urban_area.plot(
    col='time',
    figsize=(10, 5),
    robust=True,
    add_colorbar=False
);
```

```python
pixel_length = query["resolution"][1]  # in metres
area_per_pixel = pixel_length**2 / 1000**2

urban_area_km2 = urban_area.sum(dim=['x', 'y']) * area_per_pixel

# Plot the resulting area through time
fig, axes = plt.subplots(1, 1, figsize=(4, 4))
plt.bar([0, 1], 
        urban_area_km2, 
        tick_label=urban_area_km2.time, 
        width = 0.8, 
        color=['red', 'green']
       )
axes.set_xlabel("Year")
axes.set_ylabel("Built area (km$^2$)");

for y in urban_area_km2.time.values:
    print('Urban extent in '+str(y)+": "+str(round(float(urban_area_km2.sel(time=y).values),3))+' km2')
```
